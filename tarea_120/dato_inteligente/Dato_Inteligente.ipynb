```python
#primero creamos una lista en la que metemos convertido a float todos los valores de la columna que no son N.A.
TasaFertilidadGlobal=[]
for i in range(0,len(df)):
    if (df['Tasa Fertilidad'][i]) != "N.A.":
        TasaFertilidadGlobal.append(float(df['Tasa Fertilidad'][i]))
```

```python
#ahora hacemos de dos maneras diferentes el cálculo del promedio
```

```python
#primero mediante métodos de la librería statistics
import statistics
mean1=statistics.mean(TasaFertilidadGlobal)
print(mean1)
```

```python
#segundo mediante suma y división por el número de valores
mean2=sum(TasaFertilidadGlobal)/len(TasaFertilidadGlobal)
print(mean2)
```
